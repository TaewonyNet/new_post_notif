```python
# -*- coding: utf-8 -*-

from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError

import bs4
from bs4 import BeautifulSoup as bs

from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException

import difflib
from pprint import pprint
import re
```

```python
target_sites = [  
    'https://kakaopay.recruiter.co.kr/app/jobnotice/list', # 카카오페이 - Chrome dev tool 막아놓음. 
#     'https://kakaobank.recruiter.co.kr/app/jobnotice/list', # 카카오뱅크 - Chrome dev tool 막아놓음. 
]
```

```python
keywords = [
    '채용',
    '모집',
    '영입',
    '담당',
    
    '인턴',
    '신입',
    '경력',
]

keywords_re = r"|".join(keywords)
regex = re.compile(keywords_re)

regex
```

```python
driver = webdriver.Chrome()
```

```python
html_list = []

for i in target_sites:
    driver.get(i)
    delay = 3 # seconds
    my_Elem = WebDriverWait(driver, timeout=delay).until(EC.visibility_of_all_elements_located((By.TAG_NAME, "body")))
    new_driver = my_Elem[0].parent
    req = new_driver.page_source
    print(req)
    html_list.append(req)
```

```python
html_list_bs = [bs(x, 'html.parser') for x in html_list]

html2site = {}
for i in range(0, len(html_list_bs)):
    html2site[html_list_bs[i]] = target_sites[i]

len(html_list_bs)
```
